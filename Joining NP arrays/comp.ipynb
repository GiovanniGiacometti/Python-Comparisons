```python
import numpy as np
from timeit import default_timer as timer
import matplotlib.pyplot as plt
```

```python
n_trials = 1000000
dimensions_range = range(2,20)
```

### Varying dimensions

##### Concat

```python
dimensions_concat = []
for dimension in dimensions_range:
    times_concat = []
    a = np.random.randn(20,dimension)
    b = np.random.randn(1,dimension)
    for i in range(n_trials):
        t = timer()
        c = np.concatenate([a,b])
        times_concat.append(timer()-t)
    dimensions_concat.append(np.mean(times_concat))
```

##### Vstack

```python
dimensions_vstack = []
for dimension in dimensions_range:
    times_vstack = []
    a = np.random.randn(20,dimension)
    b = np.random.randn(1,dimension)
    for i in range(n_trials):
        t = timer()
        c = np.vstack([a,b])
        times_vstack.append(timer()-t)
    dimensions_vstack.append(np.mean(times_vstack))
```

#### Compare

```python
plt.plot(dimensions_concat, label = "concatenate")
plt.plot(dimensions_vstack, label = "vstack")
plt.title("Concat and vstack vs dimensions ")
plt.xlabel("Dimensions")
plt.ylabel("Execution time")
plt.grid()
plt.legend()
plt.xticks(ticks = range(len(dimensions_range)), labels = dimensions_range);
```

### Varying samples

The second array has always 1 sample.

```python
n_trials = 1000
samples_range = range(1000,20000,1000)
```

##### Concat

```python
samples_concat = []
for n_samples in samples_range:
    times_concat = []
    a = np.random.randn(n_samples,100)
    b = np.random.randn(1,100)
    for i in range(n_trials):
        t = timer()
        c = np.concatenate([a,b])
        times_concat.append(timer()-t)
    samples_concat.append(np.mean(times_concat))
```

##### Vstack

```python
samples_vstack = []
for n_samples in samples_range:
    times_vstack = []
    a = np.random.randn(n_samples,100)
    b = np.random.randn(1,100)
    for i in range(n_trials):
        t = timer()
        c = np.vstack([a,b])
        times_vstack.append(timer()-t)
    samples_vstack.append(np.mean(times_vstack))
```

#### Compare

```python
plt.figure(figsize=(20,5))
plt.plot(samples_concat, label = "concatenate")
plt.plot(samples_vstack, label = "vstack")
plt.title("Concat and vstack vs #samples ")
plt.xlabel("#Samples")
plt.ylabel("Execution time")
plt.grid()
plt.legend()
plt.xticks(ticks = range(len(samples_range)), labels = samples_range);
```
